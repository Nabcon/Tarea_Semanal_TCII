#### Grafico 2.a) Notch Pasa Bajo

```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt

num = [1/4,0, 1]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax1) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)
#MODULO
ax1.semilogx(w,mag,label='Notch LP',linewidth=3.0,color='orange')
ax1.grid(True)
ax1.set_title('Magnitude Response')
ax1.set_ylabel('Magnitude[dB]')
ax1.set_ylim([-40,1])
ax1.legend()
```

#### Grafico 2.b) Elimina Banda

```python
num = [1,-0.5*np.sqrt(2), 1]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax1) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)
#MODULO
ax1.semilogx(w,mag,label='Elimina banda',linewidth=3.0,color='orange')
ax1.grid(True)
ax1.set_title('Magnitude Response')
ax1.set_ylabel('Magnitude[dB]')
ax1.set_ylim([-10,1])
ax1.legend()
```

#### Grafico 2.c) Pasatodo

```python
num = [1,-np.sqrt(2), 1]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax2) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)

#Fase
ax2.semilogx(w, phase,label='Pasa Todo',linewidth=3.0,color='orange')
ax2.grid(True)
ax2.set_title('Phase Response')
ax2.set_ylabel('Phase[deg]')
ax2.legend()
```

#### Grafico 2.d) Notch Desplazado (Numerador)

```python
num = [1,0 , 4**2]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax2) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)

#Fase
ax2.semilogx(w, phase,label='Notch Desplazado',linewidth=3.0,color='orange')
ax2.grid(True)
ax2.set_title('Phase Response')
ax2.set_ylabel('Phase[deg]')
ax2.legend()
ax2.set_ylim([-180,30])
```
